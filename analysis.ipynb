```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
top25coins = pd.read_csv('files/top25coins.csv')

# Replace 0 with NaN
top25coins = top25coins.replace(0, np.nan)

# Convert the 'date' column to datetime
top25coins['date'] = pd.to_datetime(top25coins['date'], format='%Y-%m-%d')

# Set the 'date' column as the index
top25coins.set_index('date', inplace=True)

# Plot the data
top25coins.plot(figsize=(15, 8), title="Top 25 Coins Over Time")
plt.xlabel("Date")
plt.ylabel("Price (USDT)")
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Enable GPU acceleration if available
physical_devices = tf.config.list_physical_devices('GPU')
if physical_devices:
    try:
        tf.config.experimental.set_memory_growth(physical_devices[0], True)
        print("GPU acceleration enabled!")
    except RuntimeError as e:
        print(e)
else:
    print("No GPU found. Running on CPU.")

# Load Solana data
solana_data = pd.read_csv('files/solana.csv')
solana_data['Date'] = pd.to_datetime(solana_data['Date'], format='%m/%d/%Y')
solana_data.set_index('Date', inplace=True)

# Drop NaN values and select 'Price' column
solana_prices = solana_data[['Price']].dropna()

# Scale the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(solana_prices)

# Prepare the data for LSTM
sequence_length = 60  # Use past 60 days to predict the next day
X, y = [], []
for i in range(sequence_length, len(scaled_data)):
    X.append(scaled_data[i-sequence_length:i, 0])
    y.append(scaled_data[i, 0])
X, y = np.array(X), np.array(y)

# Reshape the data for LSTM input
X = np.reshape(X, (X.shape[0], X.shape[1], 1))

# Build the LSTM model
model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(X.shape[1], 1)),
    LSTM(50, return_sequences=False),
    Dense(25),
    Dense(1)
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X, y, batch_size=32, epochs=50)

# Predict the next 90 days
last_sequence = scaled_data[-sequence_length:]
predictions = []
for _ in range(90):
    prediction = model.predict(last_sequence.reshape(1, sequence_length, 1))
    predictions.append(prediction[0, 0])
    last_sequence = np.append(last_sequence[1:], prediction, axis=0)

# Inverse transform the predictions
predicted_prices = scaler.inverse_transform(np.array(predictions).reshape(-1, 1))

# Ensure correct last date detection
last_date = solana_prices.index.max()
if last_date.year < 2025:
    raise ValueError(f"Unexpected last_date: {last_date}. Check CSV file format.")

# Generate future dates correctly
future_dates = pd.date_range(start='2025-03-30', periods=90, freq='D')
predicted_df = pd.DataFrame({'Predicted Price': predicted_prices.flatten()}, index=future_dates)

# Save the predictions to a CSV file
predicted_df.to_csv('files/solana_predictions.csv')

# Ensure historical data is plotted until March 29th
historical_cutoff = pd.Timestamp('2025-03-29')
historical_prices = solana_prices.loc[:historical_cutoff]

# Ensure predictions start from March 30th and extend for 3 months
prediction_start = pd.Timestamp('2025-03-30')
prediction_end = prediction_start + pd.DateOffset(months=3)
predictions_to_plot = predicted_df.loc[prediction_start:prediction_end]

# Debugging: Print date ranges
print("Historical data range:", historical_prices.index.min(), "to", historical_prices.index.max())
print("Prediction data range:", predictions_to_plot.index.min(), "to", predictions_to_plot.index.max())

# Plot the predictions
plt.figure(figsize=(15, 8))
plt.plot(historical_prices.index, historical_prices['Price'], label='Historical Prices', color='blue')
plt.plot(predictions_to_plot.index, predictions_to_plot['Predicted Price'], label='Predicted Prices', linestyle='--', color='orange')
plt.title('Solana Price Prediction (March 30th Onward)')
plt.xlabel('Date')
plt.ylabel('Price (USDT)')
plt.legend()
plt.tight_layout()
plt.show()
```
